## 04. Independence and conditional probabilities

```python
bag = ['r','r','r','g','g','g','g']
experiments = 10000
counter = 0
for i in range(experiments):
    color1 = bag[random.randrange(len(bag))]
    bag.remove(color1)
    color2 = bag[random.randrange(len(bag))]
    if color1 == 'r' and color2 == 'r' :
        counter += 1
    
    bag.append(color1)

counter / experiments * 100 
```

## 05. Expected values

```python
## number of throw to get a head
outcome_space = ['H','T']
experiments = 10000
trials = 0
for i in range(experiments):
    seen_heads = False
    while not seen_heads:
        trials += 1
        result = outcome_space[random.randrange(len(outcome_space))]
        if result=='H': 
            break

trials / experiments
```

```python
## expected man wearing is hat
experiments = 1000
proper_hats = 0
men = list(range(100))
for i in range(experiments) :
    hats =list(range(len(men)))
    for man in men:
        hat = hats[random.randrange(len(hats))]
        if man == hat:
            proper_hats += 1 
            hats.remove(hat)

proper_hats / experiments
```

## 06. Discrete probability distributions

```python
experiments = 1000
experiments_k2 =0
coinflips = 10
for i in range(experiments):
    heads = 0
    for flip in range(coinflips):
        if random.random() < 0.5:
            heads += 1
    if heads == 2:
        experiments_k2+=1

experiments_k2/experiments
```

```python
experiments = 1000
sum = 0
for i in range(experiments):
    dice_1 =random.randrange(1,7)
    dice_2 =random.randrange(1,7)
    sum +=dice_1 + dice_2

sum/(experiments*2)
```

## 07. The normal distribution and the central limit theorem


```python
import matplotlib.pyplot as plt
%matplotlib inline
import random
import numpy as np

# fill a list with 100,000 random numbers between 0 and 1
random_numbers = [random.random() for i in range(100000)] 
#plot to show that numbers are not normaly distributed
plt.hist(random_numbers,bins=np.linspace(0,1,101));
```

```python
means = []
samplesize = 5
experiments = 100000
for i in range(experiments):
    total = 0
    for ii in range(samplesize):
        total += random.random()
    mean = total / samplesize
    means.append(mean)
plt.hist(means,bins=np.linspace(0,1,101));
```

## 08. Using statistical libraries

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

uniform_data = stats.uniform.rvs(
    size=10000, # volume
    loc = 1, #min value
    scale = 6 #max value
).astype(int)
uniform_data
plt.hist(uniform_data, bins=6)
plt.show()
```

```python
# know the mean of 5 throw
indexes=np.arange(0,10000,5)
sums = np.add.reduceat(uniform_data, indexes) 
np.mean(sums/5)
plt.hist(sums/5, bins=12)
plt.show()
```
